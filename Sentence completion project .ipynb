```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("qoute_dataset.csv")
```

```python
df['quote'][0]
```

# Only seperate the quotes as Scientist name isn't required

```python
quotes=df['quote']
quotes.head()
```

# 1.Converting to Lower Case

```python
quotes=quotes.str.lower()
```

# 2.Remove punctuation

```python
import string
trans=str.maketrans('', '', string.punctuation)
quotes=quotes.apply(lambda x:x.translate(trans))
```

# 3.Tokenization : convert sentemce to array then give numbers to each value according to how many times they appeared

```python
from tensorflow.keras.preprocessing.text import Tokenizer
```

# Set vocab_size first

```python
vocab_size=10000
tokenize = Tokenizer(num_words=vocab_size)
tokenize.fit_on_texts(quotes)
```

# Important

```python
word_index=tokenize.word_index
print(len(word_index))
list(word_index.items())[:10]
```

# apply the tokenized dictionary to dataset

```python
seq=tokenize.texts_to_sequences(quotes)
```

# An Example Tokenization

```python
for i in range(3):
 print(quotes[i])
```

```python
for i in range(3):
 print(seq[i])
```

# Using N grams for input nad output
# U can use for loop also

```python
x_inp = []
y_out = []

input_sequences_bigram = []
output_sequences_bigram = []

for sentence in seq:
    # Generate unigrams (single word input, next word output)
    for i in range(1,len(sentence)):
        x_inp.append(sentence[:i])
        y_out.append(sentence[i])

    # Generate bigrams (two consecutive words input, next word output)
    for i in range(len(sentence) - 2):
        input_sequences_bigram.append(sentence[i:i+2])
        output_sequences_bigram.append(sentence[i+2])

print(f"Total unigram input-output pairs: {len(x_inp)}")
print(f"Total bigram input-output pairs: {len(input_sequences_bigram)}")

print("\n--- Example for the first sentence ---")
first_sentence = seq[0]
print(f"Original tokenized sentence: {first_sentence}")

print("\nunigram examples for the first sentence:")
for i in range(len(first_sentence) - 1):
    print(f"Input: {first_sentence[i]}, Output: {first_sentence[i+1]}")

print("\nbigram examples for the first sentence:")
for i in range(len(first_sentence) - 2):
    print(f"Input: {first_sentence[i:i+2]}, Output: {first_sentence[i+2]}")
```

# Padding:Making every sequence of same length

```python
max_sequence_length = max(len(s) for s in x_inp)
print(f"The maximum length of a tokenized sentence in your dataset is: {max_sequence_length}")
```

# Make all values equal to the max length

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences
x_pad= pad_sequences(x_inp,maxlen= max_sequence_length,padding='pre')
```

# Example for first number

```python
x_pad[0]
```

```python
y_out=np.array(y_out)
```

```python
y_out.shape
```

```python
x_pad.shape
```

# one_hot encode y_out as the output will be in o's and 1's .
# Like the place where the word is will be 1 and rest 0

```python
from tensorflow.keras.utils import to_categorical
y_cat=to_categorical(y_out,num_classes=vocab_size)
```

```python
y_cat.shape
```

# Embeddings

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding,SimpleRNN,LSTM,Dense
```

```python
embed=50
rnn_units=128
```

```python
rnn_model= Sequential()
# Input Layer (Embedd layer)
rnn_model.add(Embedding(input_dim=vocab_size,output_dim=embed,input_length=max_sequence_length))
# Hiddden Layer
rnn_model.add(SimpleRNN(units=rnn_units))
# Output Layer
rnn_model.add(Dense(units=vocab_size,activation='softmax'))


```

# Optimizer

```python
rnn_model.compile(
    optimizer='adam',
    loss='categorical_crossentropy',
    metrics=['accuracy']
    )
```

```python
rnn_model.summary()
```

# LSTM Model(This can also  be used)

```python
lstm_model= Sequential()
# Input(Embed layer)
lstm_model.add(
    Embedding(input_dim=vocab_size,output_dim=embed,input_length=max_sequence_length)
)
# Middle layer
lstm_model.add(LSTM(units=rnn_units))
# Output
lstm_model.add(Dense(units=vocab_size,activation='softmax'))

```

# Compiling LSTM

```python
lstm_model.compile(
    optimizer='adam',
    loss='categorical_crossentropy',
    metrics=['accuracy']
    )
```

```python
lstm_model.summary()
```

# Training model

```python
epochs=10
batch_size=128
```

```python
history_rnn=rnn_model.fit(x_pad,y_cat,epochs=epochs,batch_size=batch_size,validation_split=0.1)
```

```python
epochs=100
batch_size=128

history_lstm=lstm_model.fit(x_pad,y_cat,epochs=epochs,batch_size=batch_size,validation_split=0.1)
```
